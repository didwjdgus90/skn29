```python
# 공유변수(클래스 변수)
# 개인, 객체변수(인스턴스 변수)
# 젤리 슬라임 키우기
# 클래스명 : Slime
# 변수 생성된 객체 수를 셀수 있는 변수 total_slimes
# 변수 슬라임을 구분하는 name color hp size

# 규칙
# 객체가 생성될 때 마다 total_slimes를 한 개씩 증가시켜서 최종 몇개의 객체가 생성되었는지
# 함수 : eat 밥먹기 --> hp이 10씩 증가, size가 5씩 증가
# 함수 : play 놀아 줄때 마다 hp가 15씩 감소 조건이 체력이 10이상이면 그외에는 밥달라고 조르기
# 함수 : 슬라임의 현재 상태를 출력( name color size hp )

class Slime:
  total_slimes = 0

  def __init__(self, name, color, hp, size):
    self.name = name
    self.color = color
    self.hp = hp
    self.size = size
    Slime.total_slimes += 1

  def eat(self,food):
    self.hp += 10
    self.size += 5
    print(f'냠냠 {self.name}이 {food}를 먹고 덩치가 커졌어요(크기 : {self.size})')

  def play(self):
    if self.hp >= 10:
      self.hp -= 15
    else:
      print("밥 주세요")

  def getStatus(self):
    print(f'이름 : {self.name}')
    print(f'색 : {self.color}')
    print(f'체력 : {self.hp}')
    print(f'크기 : {self.size}')
```

```python
a = Slime('슬라임1','파랑',100,10)
b = Slime('슬라임2','파랑',110,10)
```

```python
a.eat('사과')

a.getStatus()
```

```python
# 클래스 BankAccount 계좌
# 이름, 계좌번호, 은행명, 비밀번호, 잔고
# 예금 : 잔고에 사용자가 입력한 금액만큼 더한다
# 인출 : 사용자가 원하는 만큼의 돈을 잔고에서 인출한다.(조건문 활용.....)

class BankAccount:
  def __init__(self,accountNumber,bankname,psw,balance):
    self.accountNumber = accountNumber
    self.bankname = bankname
    self.psw = psw
    self.balance = balance
  def deposit(self, amount): #입금
    self.balance += amount
    print(f'{amount}원 입금 완료, 잔액 : {self.balance}')
  def withdraw(self,amount,psw): #출금
      if self.balance >= amount and (self.psw ==  psw) :
        self.balance -= amount
        print(f'{amount}원 출금 완료, 잔액 : {self.balance}')
      elif self.psw != psw:
        print('비밀번호가 틀렸습니다.')
      else:
        print('잔액 부족')
```

```python
ba = BankAccount("123-123",'기업','abc123',10)
ba.deposit(500000)
ba.withdraw(600000,'abc123')
```

```python
# 세탁기 WashMachine
# 변수 세탁코스 : 1 ~ 5
# 함수 오른쪽돌리기 right, 왼쪽돌리기 left
# 규칙 돌릴때마다 코스가 1씩 증가 또는 감소 5를 넘으면 1,  1보다 작으면 5
class WashMachine:
  def __init__(self):
    self.course = 1

  def turnRight(self):
    if 1 <= self.course <= 5:
      self.course += 1
    else:
      self.course = 1
    print(f'세탁코스 : {self.course}')
  def turnleft(self):
    if 1 < self.course <= 5:
      self.course -= 1
    else:
      self.course = 5
    print(f'세탁코스 : {self.course}')
```

```python
w1 = WashMachine()

for i in range(15):
  w1.turnRight()
```

```python
# 은행 번호표 TicketMachine
# 변수 : 대기번호 stand_number
# 함수 : pull_ticket : 대기번호 증가
# 함수 : reset : 초기화

# 클래스 생성
# 클래스 변수로 stand_number = 0
# pull_ticket 클래스 변수를 1씩 증가
class TicketMachine:
  stand_number = 0
  def pull_ticket(self):
    TicketMachine.stand_number += 1
    print(f'대기 번호 : {TicketMachine.stand_number}')

  def reset(self):
    TicketMachine.stand_number = 0
    print('초기화 완료')
```

```python
tk = TicketMachine()
for i in range(10):
  tk.pull_ticket()
```
