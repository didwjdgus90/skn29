```python
class Animal:
  def __init__(self, name):
    self.name = name
  def speak(self):
    print(f"동물이 소리를 냅니다.")

class Dog(Animal):
  def __init__(self,name,*bread):
    super().__init__(name)
    self.bread = bread
  def speak(self):
    print(f'{self.name}가 멍멍 짖습니다.')
  def Info(self):
    print(f'이름 : {self.name} \n품종 : {self.bread}')

class Cat(Animal):
  def __init__(self, name):
    super().__init__(name)
  def speak(self):
    print(f'{self.name}가 야옹 웁니다.')
  def Info(self):
    print(f'이름 : {self.name}')
```

```python
d = Dog("초코","포메")
c = Cat("냐옹")


d.speak()
d.Info()

c.speak()
c.Info()
```
