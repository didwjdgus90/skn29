- 생성자 :  객체를 생성할때 반드시 호출되는 내부 메소드

```python
# 붕어빵 클래스
class FishBread:
    # 생성자는 코딩하지 않으면 아래와 같이 자동으로 기본형태가 만들어져서 실행된다.
    def __init__(self):
       pass
    def myfunction(self,name=""):
        print("내가 만든 함수", name)
```

```python
# 계산기 클래스
class Calculator:
    def add(self,x,y):
      return x + y
    def minus(self,x,y):
      return x - y
    def multiply(self,x,y):
      return x * y
    def divide(self,x,y):
      return x / y
```

```python
num1 = 10; num2 = 20
cal1 = Calculator()
print(cal1.add(num1,num2))
print(cal1.minus(num1,num2))
print(cal1.multiply(num1,num2))
print(cal1.divide(num1,num2))
```

```python
cal2 = Calculator()
cal2.name = '계산기'
```

```python
cal2.name
```

```python
# 계산기 클래스
class Calculator:
    def __init__(self):
       'pass'

    def makeInfo(self,x,y):
      self.x = x
      self.y = y

    def add(self):
      return self.x + self.y
    def minus(self):
      return self.x - self.y
```

```python
c1 = Calculator()
c1.makeInfo(10,20) # 객체의 변수를 생성하는 역할(인스턴스 변수 생성)
print( c1.add(), c1.minus() )
```

```python
# 계산기 클래스
class Calculator:
    def __init__(self,x = 0 ,y = 0):
       self.x = x
       self.y = y

    def add(self):
      return self.x + self.y

    def minus(self):
      return self.x - self.y
```

```python
c2 = Calculator(20,15)
print( c2.add(), c2.minus() )
```

```python
# 클래스
# 속성 , 메소드
# 속성 : 변수(데이터)
# 메소드 : 행위(액션)

class Employee:
  # 클래스 변수(생성자 밖에서 선언)
  # 모든 직원은 Google이라는 회사에 다닌다 즉 똑같다 (공유)
  company_name = 'Google'

  def __init__(self,name,age):
     # 인스턴스 변수(생성자 안에서 선언)
     # 직원마다 이름과 나이가 다르다, 입사할때 각자 부여받는다.
     self.name = name
     self.age = age
```

```python
emp1 = Employee('홍길동',26)
emp2 = Employee('이순신',35)

print(emp1.company_name, emp1.name, emp1.age)
print(emp2.company_name, emp2.name, emp2.age)
```

```python
emp1.company_name = '삼성전자'
print(emp1.company_name, emp1.name, emp1.age)
print(emp2.company_name, emp2.name, emp2.age)
```

```python
emp3 = Employee("철이",45)
print(emp3.company_name, emp3.name, emp3.age)
```
