```python
# 부모 클래스
# 속성 : name, speed
# 함수 : drive
class Car:
  def __init__(self,name = '',speed = 0):
    self.name = name
    self.speed = speed

  def drive(self):
    print(f'{self.name} 시속 {self.speed}km로 정속 주행합니다.')
```

```python
c1 = Car('일반',60)
c1.drive()
```

```python
# 상속
class SportCar(Car):
  pass
```

```python
sc = SportCar()
sc.drive()
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def showInfo(self):
    print(f'name : {self.name} age : {self.age}')

# 오버라이딩 (over write)
class Student(Person):
  def __init__(self,name,age,score):
    super().__init__(name,age) # *** 상속 받았을 때 기본코드(중요) ***
    self.score = score
  def showInfo(self):
      print(f'name : {self.name} age : {self.age} score : {self.score}')
```

```python
# 학생객체 생성
s1 = Student("홍길동",age= 10 , score = 100)
s1.showInfo()
```
