```python
# 클래스 생성
class Dog:
  def __init__(self,name,bread,age,is_health = True):
    self.name = name
    self.bread = bread
    self.age= age
    self.is_health = is_health

  def showInfo(self):
    print(f'name : {self.name} bread : {self.bread}  age : {self.age} is_health : {self.is_health}')


# 객체생성
a1 = Dog('누렁이','불독',2,True)
a2 = Dog('바둑이','포메니안',1,True)

a1.showInfo()

# 필요하면 나중에 값을 변경
# a1.name = '누렁이'
```

```python
from typing import AsyncGenerator
# 학생 클래스 Student
# 속성(변수) 이름, 나이, 주소, 국, 영, 수
# 메소드 : 평균, 총점, 정보출력(모든 변수 다 출력)
class Student:
  def __init__(self,name, age, addr, kor, eng, math ):
    self.name = name
    self.age = age
    self.addr = addr
    self.kor = kor
    self.eng = eng
    self.math = math

  def getTotal(self):
    return ( self.kor + self.eng + self.math )

  def getAverage(self):
    return self.getTotal() / 3
```

```python
s1 = Student("홍길동",18,'서울',98,97,85)

#총점
print(f"총점 : { s1.getTotal()}")

#평균
print(f"평균 : {s1.getAverage()}")

print(s1.name,s1.age,s1.addr)
print(s1.kor,s1.eng,s1.math)
```

```python
# 쇼핑몰.. 쇼핑 카트(장바구니)
# 클래스 이름 ShoppingCart
# 변수 items 라는 리스트, 이믈
# 메소드 add_item

class ShoppingCart:
  # items = [] # --> 범인

  def __init__(self,owner):
    self.owner = owner
    self.items = []
  def add_item(self,item):
    self.items.append(item)

# 철수와 영희가 각각 빈 장바구니를 하나씩 집어 들었다.
cs = ShoppingCart("철수")
yh = ShoppingCart("영희")

# 철수가 쇼핑을 시작
cs.add_item('노트북')
cs.add_item('마우스')

# 철수의 쇼필 목록
print(f'{cs.owner} \n장바구니 {cs.items}')

# 영희가 쇼핑을 시작
yh.add_item('핸드폰')
yh.add_item('컴퓨터')

# 영희의 쇼핑 목록
print(f'{yh.owner} \n장바구니 {yh.items}')
```

```python
hg = ShoppingCart('홍길동')
#홍길동의 쇼핑 목록
print(f'{hg.owner} \n장바구니 {hg.items}')
```
