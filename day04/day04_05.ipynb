# 클래스
- 속성 : 변수
  - 인스턴스 변수
  - 클래스 변수
- 액션 : 메소드
- 상속
  - 오버라이딩

```python
class Person:
  def __init__(self,name,age):
    self.name = name
    self.age = age
  def showInfo(self):
    return f'이름 : {self.name} 나이 : {self.age}'

class Student(Person):
  def __init__(self, name, age, kor, eng, math):
    super().__init__(name,age)
    self.kor = kor
    self.eng = eng
    self.math = math
    self.total = kor + eng + math
    self.avg = self.total/3
  def showInfo(self):
    print('--학생--')
    return super().showInfo() + f' kor : {self.kor} eng : {self.eng} math : {self.math} total : {self.total} avg : {self.avg:.2f}'

class Teacher(Person):
  def __init__(self, name, age, *subjects):
    super().__init__(name,age)
    self.subjects = subjects
  def showInfo(self):
    print('--선생--')
    return super().showInfo() + f' subjects : {self.subjects} '
```

```python
# 학생
s = Student('홍길동',16,87,90,92)
print(s.showInfo())
# 선생
t = Teacher('이순신',39,'국','영','수')
print(t.showInfo())
```

```python
student_params = {
    'kor' : 100,
    'eng' : 95,
    'math' : 98
}
s2 = Student('이순신',17, **student_params) # 묶어서 사용가능
s2.showInfo()
```
