```python
# 직원정보
# person = {
#    "id" : "sk-001",
#    "이름": "김철수",
#    "나이": 35,
#    "직급": "과장",
#    "월급": 5000,
#    "스킬": ["Python", "JavaScript", "SQL"],
#    "주소": {
#        "시": "서울",
#        "구": "강남구"
#    }
#}
# 입사 - 직원 dict 생성
# 조회
# 변경
# 삭제
# 직원 정보는 파일에 저장

# 요구사항
# 1. 생성, 조회, 변경, 삭제 종료 메뉴를 제공하고 사용자가 해당 메뉴를 선택하면 적절한 로직을 수행한다.
# 2. 프로그램 시작시 파일에서 직원정보를 load 한다.
# 3. 프로그램 종료전에 수정된 직원정보를 파일에 저장한다.
```

```python
# 직원정보는 어떻게 로드 : json 구조 -dictionary
# 생성 : input을 이용해서 사용자로부터 데이터를 직접 입력 6개
# 조회 : 사용자로부터 키값을 입력 받아서 출력 - 이름으로 조회하되 id정보는 반드시 보여줌
# 변경 : id값으로 찾아서 변경한다.
# 삭제 : 조회를 해서 나온 직원중에 삭제하고자 하는 직원의 id로 삭제한다.
# 종료 : 프로그램에서 사용한 모든 리소스를 해제하고 종료 - 파일 업데이트 (덮어쓰기)
```

```python
import json

def load_data():
    employees_file = 'data/emp.json'
    try:
        with open(employees_file,'r',encoding='utf-8') as f:
            employees = json.load(f)
        return employees
    except FileNotFoundError as e:
        print(f'파일경로 확인해 주세요 : {e}')
    except Exception as e:
        print(f'알수없는 오류 관리자 문의 : {e}')


import json
def save_data(employees):
    employees_file = 'data/emp.json'
    try:
        with open(employees_file,'w',encoding='utf-8') as f:
            json.dump(employees, f, ensure_ascii=False, indent=2)  
    except FileNotFoundError as e:
        print(f'파일경로 확인해 주세요 : {e}')
    except Exception as e:
        print(f'알수없는 오류 관리자 문의 : {e}')

# 조회 프로그램
def search_emp(employee):
    keyword = input("검색할 이름 : ")
    found = True
    for emp in employee:
        if keyword in emp['name']:
            for key,value in emp.items():
                print(f"{key} : {value}")
        if not found:
            print("이름을 찾을수 없음")
        else:
            print("-"*30)
    
# 아이디 인덱스
def insert_id(employee):
    if not employee:
        return 0
    max_num = 0
    for emp in employee:
        num = int(emp['id'].split("-")[1])
        if num > max_num :
            max_num = num
    return f"sk-{max_num + 1:03d}" # 001 처럼 자리수 맞추기

# 삽입
def insert_emp(employee):
    emp = {}
    emp['id'] = insert_id(employee)
    print(f'자동 생성된 아이디 : {emp["id"]}')
    emp['name'] = input("이름 입력 : ")
    emp['age'] = input("나이 입력 : ")
    emp['rank'] = input("직급 입력 : ")
    emp['money'] = input("월급 입력 : ")

    skills = input("스킬 입력(, 구분) : ").split(',')
    emp['skill'] = [s.strip() for s in skills]

    emp['addr'] = {
        "시" : input("시 : "),
        "구" : input("구 : ")
    }

    employee.append(emp)
    print("직원 추가 완료")


# 삭제
def delete_emp(employee):
    del_id = input("삭제할 아이디 : ")

    for emp in employee:
        if emp['id'] == del_id:
            employee.remove(emp)
            print("삭제 완료")
            
# 수정
def update_emp(employee):
    emp_id = input("수정할 아이디 검색 : ")

    for emp in employee:
        if emp['id'] == emp_id:

            name = input(f'이름 : {emp["name"]}')
            if name:
                emp['name'] = name
                
            age = input(f'나이 : {emp["age"]}')
            if age:
                emp['age'] = age
            
            rank = input(f'직급 : {emp["rank"]}')
            if rank:
                emp['rank'] = rank
            
            money = input(f'월급 : {emp["money"]}')
            if money:
                emp['money'] = money
            
            print("수정완료")
            return 0
    print("ID 불일치")


# 메뉴
def showMenu():
    employee = load_data()
    try:
        while True:
            choice = int(input("1. 생성 2. 조회 \n3. 수정 4. 삭제\n5. 종료"))
            if  1<= choice <= 5:
                if choice == 1:
                    insert_emp(employee)
                elif choice == 2:
                    search_emp(employee)
                elif choice == 3:
                    update_emp(employee)
                elif choice == 4:
                    delete_emp(employee)
                elif choice == 5:
                    save_data(employee)
                    print("종료")
                    break
                else:
                    print("잘못된 입력")
    except Exception as e:
        print(f'올바른 번호를 선택하세요 {e}')

choice = showMenu()
```
