```python
# 크롤링이나 api를 이용해서 외부 데이터를 수집
# 수집된 데이터는 가공되지 않은 데이터이고 간혹 잘못된 데이터가 포함될 수 있다.
raw_data = ["100","250","알수없음","300","","450","데이터누락","500", "125.50","1.58"]
clean_data = []
error_log = []

# 비즈니스 로직 생성 시작...
for data in raw_data:
    try:
        clean_data.append( float(data) )
    except:
        error_log.append(data)

print(f'정상처리된 데이터 수 : {len(clean_data)}')
print('발견될 불량 내역')
for err in error_log:
    print("-",err)

# 예상되는 정답
# clean_data = [100,250,300,450,500,125.50,1.58]
# error_log = ["알수없음","","데이터누락"]
print(clean_data)
print(error_log)
```
