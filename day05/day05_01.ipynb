# 예외처리
- 오류가 발생시 프로그램을 종료하지않고 적절하게 처리

```python
# 정확히 정수나 문자를 혼용해서 사용하면 문제 없지만
# 소수점이 포함된 실수를 사용하면 걸러내지 못한다..

# 사용자에게 숫자를 입력받아서 + 1을 하는 로직
print('-------- 나이 입력 ---------')
input_number = input('숫자를 입력하세요')

is_number = True

for i in input_number: # 100, 100r 1t200 12.5 
   if not i.isdecimal(): # 숫자대신 문자를 잘못 입력 하였을때 + 숫자 중간에 문자가 있을때 break
        is_number = False
        break

if is_number and not input_number.count('.'):
    input_number = int(input_number)
    print(f' 입력한 숫자에 +1은 : {input_number + 1}')
else:
    print('정수만 입력하세요')
```

```python
print('-------- 나이 입력 ---------')

try: # 시도 한다.
    temp = [1,2,3]
    print(temp[3]) #indexError
    input_number = int(input('숫자를 입력하세요')) # valueError
    print('잘했어요')
except ValueError: # 예외가 발생했을때 처리 
    print('정수만 입력되어야 합니다.')
except IndexError:
    print('인덱스 범위를 체크하세요') 
```
