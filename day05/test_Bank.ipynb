```python
class BankAcoount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance

    def deposit(self,amount):
      try:
          amount = int(amount)
          if amount <= 0:
            raise ValueError('입금 금액은 0보다 커야 합니다.')
          self.balance += amount
          print(f'{amount}원 입금 되었습니다.')
      except ValueError:
        print(f"입금 오류 : 정수를 입력 해주세요")

    def withdraw(self,amount):
      try:
        amount = int(amount)
        if self.balance < 0:
            raise ValueError('잔액이 부족합니다.')
        self.balance -= amount
        print(f'{amount}원 출금 되었습니다.')
      except ValueError:
        print(f"출금 오류 : 숫자를 입력하세요")
      except Exception as e:
        print(f'출금 실패 : {e}')

    def show_balance(self):
      print(f'잔액은 {self.balance}원 입니다.')
```

```python
try:
  Bk = BankAcoount('홍길동')
  Bk.deposit(int(input('입금 할 금액을 입력하세요 :')))
  Bk.show_balance()
  Bk.withdraw(int(input('출금 할 금액을 입력하세요 :')))
  Bk.show_balance()
except Exception as e:
  print(e)
```
