```python
def calculate_average(scores):
  if type(scores) != list: # type 리스트가 아니라 딕셔너리나 튜플로 나오면 오류 출력
    raise TypeError("매개 변수는 반드시 리스트 형태가 되어야 합니다. ")
  if len(scores) == 0: # 리스트가 비어 있는 경우
    raise ValueError("리스트는 비어있으면 안됩니다.")

  return sum(scores) / len(scores)

datas = [100,'a']

try:
  print(calculate_average(datas))
except (TypeError, ValueError) as e:
  print(f"데이터 처리오류 : {e}")
except Exception as e:
  print(f'알수 없는 오류 : {e}')
  
```
