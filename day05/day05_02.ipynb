```python
try:
    temp = [1,2,3]
    # 마지막 데이터 값 출력
    # 사용자로 부터 정수를 입력받아서
    # 리스트의 마지막 값과 더한다 그리고 출력한다.
    user_input = int(input('정수를 입력하세요'))
    temp.pop()
    print(user_input + temp[2])
except: # 모든 오류를 처리하는 곳 (오류타입에 따라서 처리하지 않고)
    print("정수를 올바르게 입력하세요")
```

```python
try:
    temp = [1,2,3]
    user_input = int(input('정수를 입력하세요'))
    # temp.pop()
    print(user_input + temp[2])
    temp2 = (1,2,3)
    temp2[0] = 100 # 튜플 에러
except IndexError: 
    print("인덱스 값을 확인하세요")
except ValueError:
    print("정수 값을 확인하세요")
except:
    print("인덱스 및 정수값 에러 외 모든 에러")
```

```python
# 문제상황 : 리스트의 범위를 벗어난 데이터 요청
try:
    student_names = ['철수','영희','순희']
    req_index = int(input('조회할 학생의번호(0~2)를 입력하세요'))
    print(f'선택한 학생 : {student_names[req_index]}')
except IndexError:
    print('인덱스 값을 확인하세요')
except ValueError:
    print('올바른 값을 입력하세요')
except:
    print('관리자에게 문의하세요')
```

```python
try:
    user_input = (input('정수를 두개 공백으로 구분해서 입력(ex 10 20)'))
    number1, number2 = user_input.split() #문자열을 공백으로 나누어서 값을 number1 과 number2에 각각 저장한다. # ValueError
    number1, number2 = int(number1), int(number2) # 저장된 값을 문자열에서 숫자로 변경 한다. # ValueError
    print(number1 / number2 ) # Division by zero
except ValueError:
    print('올바른 숫자를 입력하세요')
except Exception as e: # 각 에러 클래스의 모든 것을 갖고 있음
    print(f'알 수 없는 에러 발생 {e}')
```

에러없이 코드가 실행되었을때만 작동하는 로직과, 에러 유무에 상관없이 무조건 실행되어야 하는 로직이 혼재
try ~ except ~ else

```python
# 몸무개와 키를 입력받아서 bmi 지수를 계산
try:
    weight = float(input('몸무개(kg)를 입력하세요 소수점'))
    height = float(input('키(cm)를 입력하세요 소수점'))

    if weight <= 0 or height <= 0:
        raise ValueError("키와 몸무게는 0보다 커야 합니다.") #ValueError를 발생하는 것  

    height *= 0.01 # height = height*0.01
    bmi = weight / (height**2) # 자기 자신 제곱

except Exception as e:
    print(f'에러발생 : {e}')
else: # try 항목 중 에러가 없을 때 발생한다. (if문과 다른 개념)
    print(f'계산성공! 당신의 BMI지수는 {bmi: .2f} 입니다.')
```

# try ~ except ~ else ~ final

```python
try:        # 예외가 발생할 수도 있는 코드
    pass 
    # 자원 할당 -- 성공
    # 자원 이용 -- 실패 즉 오류
    # 자원 해제
except:     # 예외 발생 시 처리
    pass 
else:       # 예외가 없을때 실행
    pass 
finally:    # 무조건 실행  
    pass 
    # 자원 해제
```

# 파일 읽기

```python
try:
    file = None
    file = open(r'C:\skn29\PYTHON\day05\test.txt', 'r') # \ -> / 바꾸거나 앞에 r로 해야됨 'r'은 읽기모드로 연거
    content = file.read()
except FileNotFoundError:
    print('파일이 존재하지 않습니다.')
else:
    print(content)
finally:
    if file is not None:
        file.close()
```

```python
# 파이썬 입장에서는 에러가 아님(문법은 ok)
# 논리적으로 오류 또는 비즈니스 관점에서 오류 함수를 사용하여 처리

# 0 < price <= 10000


def register_product(price):
    if not (0 < price <= 10000): # 값이 0 ~ 10000 사이에 값이 없으면 ValueError 출력
        raise ValueError("상품 가격은 0 ~ 10000 사이여야 합니다.")
    print(f'가격 {price}원 상품 등록 완료')

datas = [100,-100,2000,3000,0,'aaa']
for data in datas:
    try:
       register_product(data) 
    except ValueError as e:
      print(f'등록실패 : {e}')
    except Exception as e:
      print(f'알 수 없는 예외 발생 : {e}')
```
