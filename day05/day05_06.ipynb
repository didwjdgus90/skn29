로그인

아이디, 패스워드를 받아서 5번까지 반복해서 받는다.. 만약 5번 초과하면 메시지 출력하고 프로그램 중지

```python
ori_id = "USER"
ori_psw = '1234!@'

for i in range(5):
  id = input('아이디 입력 : ')
  psw = input(' 비밀번호 입력 : ')

  if id == ori_id and psw == ori_psw:
    print('로그인 성공')
    break
```

```python
# 반드시 처리 할때 사용한다.
def checkIdPassword(id,pw):
  if not (id == 'USER' and psw == "1234!@"):
    raise ValueError('아이디 또는 패스워드 정보가 잘못되었습니다.')

for i in range(5):
  try:
    id = input('아이디 입력 : ')
    psw = input(' 패스워드 입력 : ')
    checkIdPassword(id,psw)
  except Exception as e:
    print('error', e)
  else:
    print('로그인성공')
    break
```

```python
# 고객 포인트 적립 시스템
customer_db = {
    "C001": {"name": "김철수", "points": 1500},
    "C002": {"name": "이영희", "points": 3200},
    "C003": {"name": "박민수", "points": 0}
}
# 직원이 고객 아이디를 조회
id = input('고객 아이디 입력 : ')
try:
  customer_info = customer_db[id]
except:
  print('고객 정보가 없습니다.')
else:
# 추가할 포인트 값을 받아서 처리
  add_point = input(f'{id} 고객의 추가할 포인트를 입력하세요 : ')
  try:
    customer_info['points'] += int(add_point)
  except:
    print("정수만 입력 가능합니다.")
  else:
    print(f'--> 적립완료! 총 포인트 : {customer_db[id]["points"]}')

```

```python
class CustomerTable:
  customer_db = {
      "C001": {"name": "김철수", "points": 1500},
      "C002": {"name": "이영희", "points": 3200},
      "C003": {"name": "박민수", "points": 0}
  }
  def __init__(self,id):
     self.id = id

  def customeraddPoints(self):
      point = int(input('적립할 포인트 : '))
      if 0 < point < 100 :
          CustomerTable.customer_db[self.id]['points'] += point
      else:
          raise ValueError('포인트적립범위는 1 ~ 99 ')

  def customerInfo(self, id):
      print(f'{CustomerTable.customer_db[self.id]}')

# 객체 생성
try:
  id = input('고객 아이디 : ')
  cu = CustomerTable(id)
  cu.customeraddPoints()
  cu.customerInfo(id)

except KeyError as e:
    print(f'고객아디를 확인하세요 :{type(e)} {e}')
except ValueError as e:
    print(f'값을 확인하세요 : {type(e)} {e}')
except Exception as e:
    print(f'알수없는 에러 관리자 문의 : {type(e)} {e}')
```
