```python
# 각 국가별로 규격이 다르다 그리고 이미 구현한 코드에 영향을 최소화 한다.
# Phase 2
def inspect_part_v2(part_id,part_size, min_size=148.0, max_size=152.0):
    if part_size >= min_size and part_size <= max_size: 
        print(f'{part_id} : 정상 (크기 : {part_size}cm)')
    else:
        print(f'{part_id} : 불량 (크기 : {part_size}cm)')


inspect_part_v2("Bumper_AAA",150,148.0,152.0)
```

```python
# 내수용
inspect_part_v2("Bumpt_BBB", 120)
# 수출용
inspect_part_v2("Bumpt_BBB", 120, 120, 125)
```

```python
# 제품마다 방생하는 불량코드가 있음
# ERR-01 : 도색불량
# ERR-02 : 표면불량
# ERR-03 : 색상불량
# (ERR-01) (ERR-01,ERR-03), (ERR-01,ERR-02,ERR-03)
# Phase 3 가변 매개변수는 보통 가장 맨뒤 그러나 키워드 매개변수가 있으면 그 앞에
def inspect_part_v3(part_id, part_size, *errors, min_size=148.0, max_size=152.0):
    if part_size >= min_size and part_size <= max_size:
        print(f'{part_id} : 정상 (크기 : {part_size}cm)')
    else:
        print(f'{part_id} : 불량 (크기 : {part_size}cm)')
    
    if errors:
        print(f'크기불량 외 추가 불량 정보')       
        for i in errors:
            print(f'error : {i}')

inspect_part_v3("Bumpt_BBB",120,"ERR-01","ERR-02", min_size=150, max_size=145)
 
```

```python
# 품질팀에서 요구사항 : 검사할때 공장의 온도 습도 작업자 이름, 라인번호등 환경데이터도 같이 DB에 저장해주세요
# 이러한 요구사항은 매번 바뀔수 있음.. 그래서 **env_data 로 처리하기로 함
# 리펙토링 환경데이터는 *defects, **env_data
def inspect_part_final(part_id, part_size,   *defects ,  **env_data):
    print(f'{part_id} 종합 검사 로그 저장')
    # 1. 비정형 환경 데이터    
    if env_data:
        print("환경 데이터")
        for key, value in env_data.items():
            print(f' > {key} : {value}')
    if defects:
        print(f'[불량 내역] {defects}')
```

```python
inspect_part_final(
    "Fender_X9",150.5,
    "미세 균열","코팅 불량",
    factory_temp = 24.5,
    humidity="40%",
    inspector = "김엔지니어"
)
```

```python
# inspect_part_final 함수에서 min_size,max_size 어떻게 추가? 하고 테스트
def inspect_part_final(part_id, part_size,   *defects , min_size=148.0 , max_size=152.0,  **env_data): # 순서대로 ㄴ
    print(f'{part_id} 종합 검사 로그 저장')
    # 1. 비정형 환경 데이터
    if part_size >= min_size and part_size <= max_size:
        print(f'{part_id} : 정상 (크기 : {part_size}cm)')
    else:
        print(f'{part_id} : 불량 (크기 : {part_size}cm)')
    
    
    if env_data:
        print("환경 데이터")
        for key, value in env_data.items():
            print(f' > {key} : {value}')
    if defects:
        print(f'[불량 내역] {defects}')
```

```python
inspect_part_final(
    "Fender_X9",150.5,
    "미세 균열","코팅 불량",
    min_size= 150, max_size= 160,
    factory_temp = 24.5,
    humidity="40%",
    inspector = "김엔지니어"
)
```
