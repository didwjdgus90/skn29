그레고리력은 다음과 같은 예외 규칙을 추가하였다. 이에 따라 1년은 365.2425일이 된다.
100으로 나누어 떨어지되 400으로 나누어 떨어지지 않는 해는 평년으로 한다.(즉 100의 배수 중 400의 배수면 윤년이고 아니면 평년. ex) 1600(윤), 1700(평), 1800(평), 1900(평), 2000(윤), 2100(평), ...)

가령 400으로 나누어 떨어지는 2000년은 그레고리력으로 윤년이지만, 100으로는 나누어 떨어지되 400으로는 나누어 떨어지지 않는 1900년은 평년이다. 물론 율리우스력으로는 1900년이든 2000년이든 모두 윤년이다.

현재 사실상 세계표준력으로 사용되는 그레고리력에 따라 논리적으로 정리해보면 {4의 배수에서 (100의 배수가 아닌 수) 또는 (400의 배수)}인 해가 윤년이 된다.

```python
year = 2020
if (year % 4 == 0):
    print("윤년")
elif year % 100 == 0:
    print("평년")
elif year % 400 == 0 :
    print("윤년")
else:
    print("평년")
```

```python
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("윤년")
else :
    print("평년")
```

# 순환문

```python
import time # 기본 파이썬이 가지고 있지 않은 기능(라이브러리)을 추가
battery_temp = 35.0
safe_limit = 40.0

print('배터리 온도 모니터링을 시작합니다.')

while battery_temp < safe_limit:
    print(f'[정상] 현재 온도 : {battery_temp: .1f}C')
    # 실제상황에서는 값을 읽어오지만 여기서는 임의로
    battery_temp += 0.5
    time.sleep(1)

print(f'[경고] 온도가 한계치 도달 현재온도 : {battery_temp}C')
```

```python
# 연비계산기
def calculate_efficency(distance, fuel_used):
    # 주행거리를 / 사용한 연료량으로 나눔
    eff = distance / fuel_used
    return f"연비 : {eff : .3f} km/L"
```

```python
print(calculate_efficency(300,20))
```

```python
print(calculate_efficency(20,300))
```

```python
def choice_menu():
    print('메뉴')
    print('1. 입력  3. 조회')
    print('2. 출력  4. 종료')
    choice = int(input('번호를 선택하세요'))
    return choice

while True:
    choice = choice_menu()
    if choice == 1:
        print('정보를 입력하세요')
    elif choice == 2:
        print('사용자 정보 시스템 입니다.')
    elif choice == 3:
        print('조회할 번호를 입력하세요. ')
    elif choice == 4:
        print('프로그램을 종료합니다.')
        break


```

# 가변 매개변수
- 입력된 값의 개수를 미리 알 수 없을 때 사용 이때는 변수이름 앞에 *붙여서 구분 
- 내부에서는 전달된 값들이 튜플로 변환된다. -- 패킹

```python
# N개 센서의 오차율 평균 구하기
def get_average_error(*erros):
    if not erros:
        return 0
    total = sum(erros)
    return total / len(erros)

get_average_error(1,2)
```

```python
# 주행로그를 처리하는 함수
def sae_driving_log(vehicle_id, **log_data):
    print(f'-------- 차량 {vehicle_id} 주행로그 ----------')
    for key, value in log_data.items():
        print(f'{key} : {value}')

sae_driving_log('CAR-01',weather = '비', temperature = 22, speed = 120)
```

```python
# 가변매개변수 
def funn1(model, *trains) :
    print(f'학습할 모델명 : {model}')
    print(f'학습에 사용하는 데이터의 개수는 : {len(trains)}')
    for i in trains:
        print(f'개별데이터 : {i} ')
```

```python
funn1('Tree', 10,20,30,40)
```

```python
def funn2(model, **hyper_params):
    print(f'사용한 모델: {model}')
    print(f'모델의 성능을 향상시키기위한 하이퍼 파라메터는')
    for key, value in hyper_params.items():
        print(f'{key} : {value}')
        
funn2('Tree', epochs = 100, lr = 0.02, batch_size = 256)
```
