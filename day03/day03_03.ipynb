```python
# 함수이름 dispatch_robotaxi
# 자율주행 로보택시
# 관제센터로 들어오는 운행기록(log) 제각각, 혼자타는 사람, 5명 탑승, 비오는날, 배터리가 없는날, 센서가 고장
# 모든 변화무쌍한 변수를 하나의 함수로 통제하는 배차시스템 함수
# 1. 위치 : 차량번호, 목적지(배차에 절대 빠지면 안되는 필수 정보)
# 2. 디폴트 : 운행모드(설정안하면 컴포트모드로 운행)
# 3. 가변(*args) : 탑승객 명단(빈 차일수도 있음)
# 4. 키워드 가변(**kwargs) : 주행환경 및 센서데이터(어떤 데이터가 추가될지 예측불허)
# 평일
# dispatch_robotaxi('TAXI-001','강남역','컴포트','홍길동')
# 금요일 저녁 다수의 탑승객
# dispatch_robotaxi('TAXI-002','홍대입구','스포츠','홍길동','빅데이터','머신러닝')
# 최악의 상황.. 배터리도 다되어가고, 비도오고 차량정체 심하고, 센서오류
# dispatch_robotaxi('TAXI-003','인천공항','에코모드','VIP대표','수행비서', weather='폭우', battery_noise='Hight',
#  wiper_speed='MAX', sensor_error = 'Error-002')
```

```python
def dispatch_robotaxi(car_id, car_distance,  args , car_mode = "컴포트" ,  **kwargs):
    print(f' 차량 번호 : {car_id}\n 목적지 : {car_distance}')
    print(f' 차량 모드 [{car_mode}]')
    if args:
        print(f" [탑승자 명단] : {args}")
    if kwargs:
        print("--주변환경 및 센서데이터--")
        for key, value in kwargs.items():
            print (f' {key} : {value} ')

dispatch_robotaxi(
    'TAXI-001','강남역',
    '홍길동'
)
```

```python
def dispatch_robotaxi(car_id, car_distance,  *args , car_mode = "컴포트" ,  **kwargs):
    print(f'차량 번호 : {car_id}\n목적지 : {car_distance}')
    print('탑승객 명단')
    for i in args:
        print(i,end=' ')
    print(f'\n차량 모드 [{car_mode}]')
    print('---주행환경(에러포함)---')
    for key ,value in kwargs.items():
        print(f'{key} : {value}')
```
