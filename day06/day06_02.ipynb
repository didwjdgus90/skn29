```python
# 자신의 이름과 나이 취미를 포함한 간단한 자기소개 파일을 만들어 주세요.
# 요구사항
# 파일명 : introduction.txt
# 3줄이상
# utf-8
# 모드는 w
with open('introduction.txt','w', encoding="utf-8") as f:
    f.write('-----자기소개서------\n')
    f.write('이름 : 양정현\n')
    f.write('나이 : 26\n')
    f.write('취미 : 런닝\n')
    f.write('--------------------')
```

```python
# number.txt 숫자를 읽어서 합 평균 최대값 최소값을 계산 -> 기초 통계량
# 요구사항 readline() 또는 반복문을 이용 숫자로 변환
numbers=[]
with open('numbers.txt','r', encoding = 'utf-8') as f:
    for line in f:
        numbers.append(int(line.strip())) # numbers 안에 int형태로 추가한다.
print(f'sum : {sum(numbers)}')
print(f'avg : {sum(numbers) / len(numbers)}')
print(f'max : {max(numbers)}')
print(f'min : {min(numbers)}')
    
```

```python
# poem.txt 파일을 읽어서 라인 번호를 넣어서 출력
# 요구사항 enumerate(temp,1)
# [1] 산 넘어 산
# [2] 물 건너 물
# [3] 나 혼자 가는 길
with open('poem.txt','r',encoding='utf-8') as f:
    line = f.readlines()

with open('poem.txt','w',encoding= 'utf-8')as f:
    for index, value in enumerate(line,1):
        f.write(f'[{index}] {value.strip()} \n')
```

```python
# part1 ~ part3 합쳐서 하나의 파일로 생성 merged.txt
# 요구사항
# 3개의 파일을 순서대로 읽기
# 파일 사이에 구분선 ----------------------추가 print('-'*30)
# 병합 후 저장
```

```python
files = ['datas\part1.txt','datas\part2.txt','datas\part3.txt']
with open('merged.txt','w', encoding = 'utf-8') as f:
    for fname in files:
        with open(fname,'r',encoding = 'utf-8') as fs:
            f.write(fs.read())
            f.write("-"*30)
            f.write("\n")
```

```python
with open('merged.txt','r', encoding = "utf-8") as f:
    print(f.read())
```

```python
import glob
with open('merged.txt','w', encoding="utf-8") as meged_f:
    file_list = glob.glob(r'datas\*.txt')
    for index, filepath in enumerate(file_list):
        with open(filepath, 'r') as f:
            meged_f.write(f.read())
            if index != len(file_list) -1:
                meged_f.write('-'*30 + '\n')
```
