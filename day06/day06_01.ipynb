# 파일 읽는 방법
- 1. 전체 내용을 한번에 가져오기
- 2. 한줄씩 읽기
- 3. 모든 줄을 리스트에 읽기
- 4. 반복문을 이용해서 줄 단위로 읽기 --> 가장 추천되는 방법

```python
# 1. (크기많아 죽을수도 있다.)
with open('text.txt','r') as f:
    content = f.read()
    print(content)
```

```python
# 2. (다 읽을려면 while문 사용)
with open("text.txt",'r') as f:
    print(f.readline())
    print(f.readline())
```

```python
# 3.
with open('text.txt','r') as f:
    lines =  f.readlines()
    print(lines)
    print(type(lines))
```

```python
# 4.
with open('text.txt','r') as f:
    for line in f: # 한줄씩 반복
        print(line.strip()) #\n을 제거하고 출력한다.
```
