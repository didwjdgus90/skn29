# csv란
- 쉼표로 구분, 액셀로 열 수 있고, 데이터베이스에 자주 나타남, 가볍고 호환성이 좋다.
  - 이름, 나이, 직급, 월급
  - 김철수,35, 과장, 5000
  - 이영희,28, 대리, 3500
  - 박민준,42, 부장, 7000

```python
import csv

with open('employes.csv', 'w', encoding='utf-8', newline='') as f:
    writer = csv.writer(f)
    # 헤더 행
    writer.writerow(['이름', '나이', '직급', '월급'])
    # 데이터 행
    writer.writerow(['김철수', 35, '과장', 5000])
    writer.writerow(['김철수', 28, '과장', 3500])
    writer.writerow(['김철수', 42, '과장', 7000])
```

```python
# 읽기
import csv
with open('employes.csv','r', encoding='utf-8') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row, type(row))
```

```python
# 읽기
import csv
with open('employes.csv','r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row, type(row))
```

```python
import pandas as pd
# 데이터 생성
data = {
    '이름': ['김철수', '이영희', '박민준'],
    '나이': [35, 28, 42],
    '직급': ['과장', '대리', '부장'],
    '월급': [5000, 3500, 7000]
}
df = pd.DataFrame(data)
df.to_csv('employ2.csv', encoding='utf-8', index= False)
```

```python
person = {
    "이름": "김철수",
    "나이": 35,
    "직급": "과장",
    "월급": 5000,
    "스킬": ["Python", "JavaScript", "SQL"],
    "주소": {
        "시": "서울",
        "구": "강남구"
    }
}
import json
# 파일로 json 객체정보를 저장
with open('person.json','w',encoding='utf-8') as f:
    json.dump(person,f,ensure_ascii=False, indent=2) #person.json을 보내다 #indent 들여쓰기 # 문자열로 출력
```
