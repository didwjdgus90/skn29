데이터 타입과 조건문(if) 결합
- if, elif, else
- Truthy, Falsy : bool 타입이 아니여도 조건문에서 참/거짓으로 평가되는 값들이 있음
- 숫자 0, 빈 문자열 "", 빈리스트 [] 등은 False로 판단

```python
status_coe = 200
error_message = ""

if status_coe == 200:
    print("데이터 수집 성공")
elif status_coe == 400:
    print("페이지를 찾을수 없음")
else :
    print("알 수 없는 오류")

if  not error_message:
    print("에러 로그가 비었씁니다. 정상 작동 중.")
```

논리연산
- and : 조건이 하나라도 False 이면 전체가 False ~이고 ~이다 싸고 맛있는 집
- or : 조건이 하나라도 True 이면 전체가 True ~ 또는 싸거나 맛있는 집
- not

```python
score = 80
# 점수가 80이상이고 90점 미만
is_valid = (score >= 80) and (score < 90)
print(f"조건 만족 여부 : {is_valid}")
```

```python
is_admin = False
is_author = True

# 관리자나 작성자에게 수정 권한
can_edit = is_admin or is_author
print(f'수정 권한 있음 : {can_edit}')
```

```python
# 수집된 베터리 센서 데이터
battery_temp = 75.5
battery_volatage = 11.2
is_charging = True

#1. 정상조건 : 온도가 80도 미만이고, 전압이 12v 이상이어야 정상
if (battery_temp < 80) and (battery_volatage >= 12) :
    print("정상")
#2. 온도가 90도 이상이거나, 전압이 10v 미만이면 경고 발생
elif (battery_temp >= 90) or (battery_volatage < 10) :
    print ("경고")
#3. 유지보수필요(복합연산) 충전중이 아닌데, 전압이 11.5v 미만이면 점검필요.
elif (not is_charging) and (battery_volatage < 11.5) :
    print ("점검필요")
else :
    print ("비정상")
```

자료구조 : 데이터를 담는 바구니
- 리스트 와 딕셔너리


```python
# 순서가 있는 데이터 묶음( 센서 데이터 저장)
sensor_logs = [25.4 , 26.1, 24.8 , 20.5] # create
sensor_logs.append(22.2) # create
# read 26.1 출력
print(sensor_logs[1])
# update
sensor_logs[1] = 22.5
print(sensor_logs)
# delete index -1
print(sensor_logs.pop())
print(sensor_logs)
sensor_logs.remove(22.5)
# 저장순서를 보장 최근 2개를 출력
print(sensor_logs[-2:])
```

딕셔너리 {}
- key : value
- key는 중복안됨, value는 제한 없음
- 주민등록번호 : 이름

```python
device_config = {
    'device_id' : 'N-4090',
    'location' : 'A-Block',
    'status' : 'Online',
    'sampling_rate' : 0.5
}
student_config = {
    'name' : '홍길동',
    'kor' : 100,
    'eng' : 95,
    'math' : 100,
    'addr' : '서울시 강남구'
}

# read
print(f"device id : {device_config['device_id']}")
# update 학생이름은 홍길동 -> 길동이
student_config['name'] = '길동이'
print(student_config)
# delete
print(student_config.pop('name'))
print(student_config)
# 추가
student_config['name'] = '철이'
print(student_config)
student_config.items()
```
