#파이썬 자료형 개요 및 숫자형
- 자료형(Data Type) : 컴퓨터가 데이터를 처리학위해서 종류를 구분하는 기준
- 정수(int)와 실수(float)
- 연산자 : + - * / ,  //(몫), %(나머지), **(제곱)

```python
batch_size = 32  # 정수형(int)
learning_rate = 0.001   # 실수형(float)
```

사칙연산자 및 나머지 그리고 몫 연산자

```python
batch_size = 32
total_data = 100
result = total_data + batch_size # +
print(total_data + batch_size)
# -
print(total_data - batch_size)
# *
print(total_data * batch_size)
# /
print(total_data / batch_size)   # 3.125
# //
print(total_data // batch_size) # 3
# %
print(total_data % batch_size)  # 4
# **
print(total_data ** batch_size)

```

```python
batch_size = 32
total_data = 100
result = total_data + batch_size
print(result)
result = total_data - batch_size
print(result)
```

```python
batch_size == total_data
```

불(boolean) 과 논리 연산
- True, False만 나타내는 자료형
- 비교연산자 ==(같다)  !=(다르다) <, >, >=, <=
- 논리연산자 : and(둘다 참일때 참), or(둘중 하나라도 참이면 참), not(참/거짓 반전)

```python
sensor_value = 85
threshold = 90  # 임계치
# 비교연산자
is_warning = sensor_value >= threshold
print("경고상태 인가요?",is_warning)
print(f"경고상태 인가요? {is_warning}")
```
